# Pandas and Seaborn

## 1. Importing pandas

```python
import pandas as pd
print(pd.__version__)
```

## 2. Creating a Pandas series from a NumPy array

```python
import numpy as np
a = np.array(range(10,20))
print(a)
s = pd.Series(a)
print(s)
```

## 3. Converting series index to dataframe column 

```python
s = pd.Series(np.random.randint(0,100,10))
df = pd.DataFrame(s,columns=["Value"])
df['Index'] = s.index # Add a new column to dataframe with Index values from series
df
```

## 4. Seaborn 

```python
import seaborn as sns
sns.get_dataset_names() # List available datasets in seaborn
df = sns.load_dataset("mpg") # Loads "mpg" dataset as a dataframe
df
```

## 5. List countries of origin in mpg dataset 

```python
# Lists unique values in 'origin' column
df['origin'].unique()
```

## 6. Extract dataframe by values of a column 

```python
# Filters dataset by rows whose 'origin' column value is 'usa'
df[df['origin']=='usa']
```
